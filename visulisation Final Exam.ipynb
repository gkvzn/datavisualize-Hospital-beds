```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(style="darkgrid")

```

```python
In_stat= pd.read_csv('country bed.csv')
```

```python
In_stat.head()
```

```python
Year= In_stat['Year']
stats= sns.lineplot(x='Year',y='Value',hue='Country',data=In_stat,palette="Set1")
plt.xticks(In_stat['Year'])
plt.rcParams["figure.figsize"] = (22,3)
plt.show()
# comparing neighboring countries of germany by visulizing the hospital bed statistics from the year 2000 to 2019 
```

```python
sns.pairplot(data=In_stat,hue= 'Country')
plt.show()
```
