```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(style="darkgrid")

```

```python
In_stat= pd.read_csv('country bed.csv')
```

```python
In_stat.head(6)
```

```python
sns.set_style('darkgrid')
Year= In_stat['Year']
stats= sns.lineplot(x='Year',y='Value',hue='Country',markers=["o", "x","*",'^',"v",">"],data=In_stat,palette="Set1")
plt.xticks(In_stat['Year'])
plt.rcParams["figure.figsize"] = (24,3)
plt.xlabel('Year')
plt.ylabel('Total beds per 1000 population')
plt.legend(loc=(1.025, 0.5))
plt.show()
# comparing Germany and neighboring countries of germany by visulizing the hospital bed statistics from the year 2000 to 2019 
```

```python
Year= In_stat['Year']
stats= sns.lmplot(x='Year',y='Value',hue='Country',data=In_stat,markers=["o", "x","*",'^',"v",">"], palette="Set1", height=7, aspect=1.8)
plt.xticks(In_stat['Year'])
plt.xlabel('Year')
plt.ylabel('Total beds per 1000 population')

plt.show()
#visualizing the hospital beds in germany and countries neighboring countries of germany as a lmplot
```

```python
sns.pairplot(data=In_stat,hue= 'Country')
plt.xlabel('Year')
plt.ylabel('Total beds oer 1000 population')
plt.show()

#visualizing the hospital beds in germany and countries neighboring countries of germany as a pairplot 
```

```python
recovery_rate = pd.read_csv('recovery.csv')
recovery_rate.head (2)
```

```python
# setting style for seaborn plot
sns.set(style="whitegrid", font="CMU Serif")
# creating a color palette 
cplt = sns.cubehelix_palette(n_colors=3)
```

```python

f, (ax1, ax2) = plt.subplots(ncols=1, nrows=2,
                             sharex=True)

ax1 = sns.barplot(x='Year', y='Value',
                  hue='Variable', data=recovery_rate, palette=cplt, ax=ax1)

# we basically do the same thing again for the second plot
ax2 = sns.barplot(x='Year', y='Value',
                  hue='Variable', data=recovery_rate, palette=cplt, ax=ax2)

#  setting the limits for the individual y axis
ax1.set_ylim(20000, 26000)
ax2.set_ylim(0, 1500)
# the upper part does not need its own x axis as it shares one with the lower part
ax1.get_xaxis().set_visible(False)
ax1.set_ylabel("")
ax2.set_ylabel("")
#  setting a new label on the plot 
f.text(0.05, 0.55, 'Discharges per 100000 population', va='center', rotation='vertical')
# remove both default legends first
ax1.get_legend().remove()
ax2.get_legend().remove()
ax2.legend(loc=(1.025, 0.5), title="Variable")
ax1.xaxis.tick_top()
ax2.xaxis.tick_bottom()
f.subplots_adjust(left=0.15, right=0.85, bottom=0.15, top=0.85)
plt.rcParams["figure.figsize"] = (22,23)
plt.show()
# this bar graph is plotted to show the total hospital beds rate with the total discharge rate in germany per 100000 population 
# from 2000 to 2019
# i used this with broken axis because these two variables has a huge value difference 
```
